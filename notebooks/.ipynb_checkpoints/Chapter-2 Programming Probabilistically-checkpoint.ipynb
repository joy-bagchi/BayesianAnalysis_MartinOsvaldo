```python
import numpy as np
import pymc as pm
import arviz as az
import preliz as pz
import matplotlib.pyplot as plt
```

```python
np.random.seed(1234)
trials = 4
theta_real = 0.35 # this will be unknown in a real-world experiment 
data = pz.Binomial(n=1, p=theta_real).rvs(trials)
data
```

```python
with pm.Model() as model:
    theta = pm.Beta('theta', alpha=1, beta=1)
    y = pm.Binomial('y', n=1, p=theta, observed=data)
    trace = pm.sample(1000)
```

```python
pm.plot_trace(trace)
```

```python
az.plot_trace(trace, kind='rank_bars', combined=True)
```

```python
az.summary(trace).round(2)
```

# Different ways of interpreting the posterior to make decision

## Savage-Dickey Density ratio

```python
az.plot_bf(trace, var_name='theta', prior=np.random.uniform(0, 1, 10000), ref_val=0.5)
```

## ROPE (Region of Practical Equivalence)

```python
az.plot_posterior(trace, rope=[0.45, 0.55])
```

## Plot the Reference Value

```python
az.plot_posterior(trace, ref_val=0.5)
```

# Loss Functions 

In real life we do not know the true value of the parameter we are trying to estimate. Hence we pic the parameter \'theta' by minimizing the loss function across the posterior distribution. In other words we pick a value of \'theta' that has the lowest value for the loss function

```python
grid = np.linspace(0, 1, 200)
theta_posterior = trace.posterior['theta']
loss_abs = [np.mean(abs(i-theta_posterior)) for i in grid]
loss_quadratic = [np.mean((i-theta_posterior)**2) for i in grid]
for loss, c in zip([loss_abs, loss_quadratic], ['Green', 'Blue']):
    mini = np.argmin(loss)
    plt.plot(grid, loss, c)
    plt.plot(grid[mini], loss[mini], 'o', color=c)
    plt.annotate('{:.2f}'.format(grid[mini]),
                 (grid[mini], loss[mini] + 0.03), color=c)
    plt.yticks([])
    plt.xlabel(r'$\hat \theta$')
plt.savefig('B11197_02_05.png')
```
